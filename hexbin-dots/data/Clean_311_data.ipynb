```python
import pandas as pd
```

```python
tree_df = pd.read_csv('311_Service_Requests_-_Tree_Debris.csv')
sanit_df = pd.read_csv('311_Service_Requests_-_Sanitation_Code_Complaints.csv')
pot_df = pd.read_csv('311_Service_Requests_-_Pot_Holes_Reported.csv')
graff_df = pd.read_csv('311_Service_Requests_-_Graffiti_Removal.csv')
print(tree_df.shape)
print(sanit_df.shape)
print(pot_df.shape)
print(graff_df.shape)
```

```python
tree_df.head()
```

```python
tree_coords = tree_df[['Latitude', 'Longitude']].copy()
tree_coords.dropna(inplace=True)
tree_coords.shape
```

```python
tree_coords['category'] = 'tree_debris'
tree_coords.rename(columns={'Latitude': 'lat', 'Longitude': 'lon'}, inplace=True)
tree_coords.head()
```

```python
sanit_coords = sanit_df[['Latitude', 'Longitude']].copy()
sanit_coords.dropna(inplace=True)
sanit_coords['category'] = 'sanitation'
sanit_coords.rename(columns={'Latitude': 'lat', 'Longitude': 'lon'}, inplace=True)
sanit_coords.head()
```

```python
pot_coords = pot_df[['LATITUDE', 'LONGITUDE']].copy()
pot_coords.dropna(inplace=True)
pot_coords['category'] = 'pot_holes'
pot_coords.rename(columns={'LATITUDE': 'lat', 'LONGITUDE': 'lon'}, inplace=True)
pot_coords.head()
```

```python
graff_coords = graff_df[['Latitude', 'Longitude']].copy()
graff_coords.dropna(inplace=True)
graff_coords['category'] = 'graffiti'
graff_coords.rename(columns={'Latitude': 'lat', 'Longitude': 'lon'}, inplace=True)
graff_coords.head()
```

```python
call_311_df = pd.concat([tree_coords, sanit_coords, pot_coords, graff_coords])
call_311_df.head()
```

```python
call_311_df.to_csv('311_calls.csv', index=False)
```
